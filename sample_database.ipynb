```python
# Standard package
import pandas as pd

# Original package
from finance.utility import Utility as util
from finance.sqlite import StockSqlite
from finance.stock_viewer import StockViewer
```

```python
# Get stock data
STOCK_DATA_DATABASE = '/Users/y/dev/oppai-main/finance/master/stock_data.db'
db = StockSqlite(STOCK_DATA_DATABASE)
df_day = db.get_data('6902.T')

# Convert
df_week = util.toWeekly(df_day)
df_month = util.toMonthly(df_day)

# Calculate indicators
indicators = {"ema":[50,100,200]}
# indicators = {"sma":[50,100,200],"ema":[50,100,200]}
df_day = pd.concat([df_day, util.calcIndicator(df_day["close"],indicators)], axis=1)
df_week = pd.concat([df_week, util.calcIndicator(df_week["close"],indicators)], axis=1)
df_month = pd.concat([df_month, util.calcIndicator(df_month["close"],indicators)], axis=1)
```

```python
StockViewer(df_day, drop_breaks=True).show()
StockViewer(df_week, drop_breaks=False).show()
StockViewer(df_month, drop_breaks=False).show()
```
